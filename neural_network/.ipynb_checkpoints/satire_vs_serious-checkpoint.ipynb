# Satire vs Serious #

This dataset is made up of tweets that link to news articles that are purposely satirical and tweets that link to serious news articles that are serious. The serious news articles are from the Wall Street Journal, labelled 0, which is seen as centrist, and satirical articles from The Onion and Babylon Bee, labelled 1. The set contains about 15,000 tweets.

```python
# import required libraries
import tensorflow as tf
from tensorflow import keras
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# choose the dataset to train the model on
DATASET = 'satire_vs_serious'

# this model uses eager execution
# clearing the session can increase training speed
tf.compat.v1.enable_eager_execution()
tf.keras.backend.clear_session()
```

```python
# Readlines into list of tweets and labels
with open('..\\dataset_creation\\datasets\\'+DATASET+'\\tweets.txt', encoding="utf8") as f:
    raw_tweets = f.readlines()

with open('..\\dataset_creation\\datasets\\'+DATASET+'\\labels.txt')as f:
    raw_labels = f.readlines()

# Create list of output targets
target_dataset = list()
for label in raw_labels:
    if label == '1\n':
        target_dataset.append(1)
    else:
        target_dataset.append(0)

# calculate 10% of size of the dataset used for splitting test and validation sets
SPLIT = round(len(raw_tweets)*0.1)
```

Function to conduct the required preprocessing and convert the tweets and target labels to tensor objects padded with a default value to make all tensors the same size

```python
def preprocess(X_batch, y_batch):
    X_batch = tf.strings.substr(X_batch, 0, 240)
    X_batch = tf.strings.regex_replace(X_batch, b"<br\\s*/?>", b" ")
    X_batch = tf.strings.regex_replace(X_batch, b"[^a-zA-Z']", b" ")
    X_batch = tf.strings.split(X_batch)
    y_batch = tf.convert_to_tensor(y_batch, dtype=tf.int64)
    return X_batch.to_tensor(default_value=b"<pad>"), y_batch
```

Preprocess the dataset and split 10% of for later testing.

```python
X_batch, y_batch = preprocess(raw_tweets, target_dataset)
```

```python
X_test, X_train_full = X_batch[:SPLIT], X_batch[SPLIT:]
y_test, y_train_full = y_batch[:SPLIT], y_batch[SPLIT:]
```

Create a vocabulary set based on the dataset and create a truncated vocabulary set of the 10,000 most common words.

```python
vocabulary = Counter()
for tweet in X_train_full:
    vocabulary.update(list(tweet.numpy()))
```

```python
vocab_size = 10000
truncated_vocabulary = [
    word for word, count in vocabulary.most_common()[:vocab_size]
]
```

Create a look up table of the 10,000 most common words and 1,000 out of value buckets

```python
words = tf.constant(truncated_vocabulary)
word_ids = tf.range(len(truncated_vocabulary), dtype=tf.int64)
vocab_init = tf.lookup.KeyValueTensorInitializer(words, word_ids)
num_oov_buckets = 1000
table = tf.lookup.StaticVocabularyTable(vocab_init, num_oov_buckets)
```

Define function that takes a matrix of tweet tensors and returns a matrix of tweets with corresponding index values inpaplace of words. Pass the training set through the function and split off a validation set.

```python
def encode_words(dataset_input):
    return table.lookup(dataset_input)
```

```python
X_valid, X_train = encode_words(X_train_full[:SPLIT]), encode_words(X_train_full[SPLIT:])
y_valid, y_train = y_train_full[:SPLIT], y_train_full[SPLIT:]
```

## Model 1 ##
* Two hidden layers
* 128 nodes
* Default activation function (tanh)

Define network layers and set embedding layer size to size of tweet tensors. Mask zero is set to true to ignore the most common word which will be the padding token.

```python
embed_size = 128
model1 = keras.models.Sequential([
    keras.layers.Embedding(vocab_size + num_oov_buckets, embed_size, mask_zero=True),
    keras.layers.GRU(embed_size, return_sequences=True),
    keras.layers.GRU(embed_size),
    keras.layers.Dense(1, activation="sigmoid")
])
```

Compile the modile call the summary function to show the layers and parameters.  Run the network for 10 epochs.

```python
model1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model1.summary()
```

```python
history1 = model1.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_valid, y_valid))
```

Represent the history from training the model in graph to show how well the model is generalising.

```python
# Represent learning curves using pandas and matplotlib
pd.DataFrame(history1.history).plot(figsize=(10, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.gca().set_xlim(0, 9)
plt.show()
```

## Evaluate and Predict ##
Model 7 shows the greatest convergence of loss and accuracy between the training and validation data.  It is still showing signs of slight over fitting. When larger datasets are run on the model the over fitting could be reduced further.

Evaluate model 7 on the test data and print the first 10 predictions with the actual labels for comparision

```python
X_test_encoded = encode_words(X_test)
```

```python
model7.evaluate(X_test_encoded, y_test)
```

Show the outcome of the first ten predictions to compare with the reloaded saved model.

```python
X_new = X_test_encoded[:10]
y_proba = model7.predict(X_new)
for i in range(len(y_proba)):
    print(str(i+1) + ".\nPrediction: " + str(y_proba[i][0].round(2)) + "\nActual: " + str(y_test[i].numpy()) + "\n")
```

Save the model and then load and predict the first ten from the test set again to ensure the model has saved correctly.

```python
model7.save("saved_models\\" + DATASET + '.h5')
```
